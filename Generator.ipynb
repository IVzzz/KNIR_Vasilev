```python
import pandas as pd
import numpy as np
df1 = pd.read_csv('ddos.csv')
df2 = pd.read_csv('norm.csv')
df1 = df1.sample(1000)
df2 = df2.sample(300)
df = pd.concat([df1, df2], ignore_index=True, sort=False)
df = df.sample(frac=1, random_state=101).reset_index()
df.fillna(0, inplace=True)
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(how="all", inplace=True)
```

```python
from sdv.single_table import GaussianCopulaSynthesizer
from sdv.metadata import SingleTableMetadata

metadata = SingleTableMetadata()
metadata.detect_from_dataframe(data=df)

model = GaussianCopulaSynthesizer(metadata=metadata)
model.fit(df)
sample = model.sample(100)
sample.head()
```

```python
from sdv.evaluation.single_table import evaluate_quality

quality_report = evaluate_quality(df,sample,metadata)
```
